```python
from bs4 import BeautifulSoup
import requests
```

```python
from csv import writer
```

```python
url="https://finance.yahoo.com/cryptocurrencies/"
User={"User-Agent": "Chrome/104.0.5112.102"}
GetData=requests.get(url,headers=User)
print(GetData)
```

```python
soup=BeautifulSoup(GetData.content,'html.parser')
```

```python
Lists=soup.find("div",{"class":"Ovx(a) Ovx(h)--print Ovy(h) W(100%)"})
print(Lists)
```

```python
with open('ikea.csv', 'w', encoding='utf8', newline='') as w:
    thewriter = writer(w)
    header = ['Name','Price','Change','Percentage,','MarketCap','Volume Since','Volume oneday','TotalVolume','Supply']
    thewriter.writerow(header)
    for l in Lists:
        rows=l.find_all('tr',{"class":"simpTblRow Bgc($hoverBgColor):h BdB Bdbc($seperatorColor) Bdbc($tableBorderBlue):h H(32px) Bgc($lv2BgColor)"})
        for i in rows:
            print(i.get_text(separator="|"))
            Name=i.find('td',{"class":"Va(m) Ta(start) Px(10px) Fz(s)"}).get_text()
            Price=i.find('td',{"class":"Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)"}).get_text()
            Change=i.find('td',{"class":"Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)"}).get_text()
            Percentage=i.find('td',{"class":"Va(m) Ta(end) Pstart(20px) Fw(600) Fz(s)"}).get_text()
            MarketCap=i.find('td',{"class":"Va(m) Ta(end) Pstart(20px) Fz(s)"}).get_text()
            VolumeSince=i.find('td',{"class":"Va(m) Ta(start) Px(10px) Fz(s)"}).get_text()
            Volumeoneday=i.find('td',{"class":"Va(m) Ta(start) Px(10px) Fz(s)"}).get_text()
            TotalVolume=i.find('td',{"class":"Va(m) Ta(start) Px(10px) Fz(s)"}).get_text()
            Supply=i.find('td',{"class":"Va(m) Ta(start) Px(10px) Fz(s)"}).get_text()
            header=[Name,Price,Change,Percentage,MarketCap,VolumeSince,Volumeoneday,TotalVolume,Supply]
            print(header)
            thewriter.writerow(header)
        
        
```
