```python
#Use urlopen and BeautifulSoup for opening and parsing
```

```python
url="http://www.hubertiming.com/results/2017GPTR10K"
html=urlopen(url)
soup=BeautifulSoup(html,'lxml')
type(soup)
```

```python
#get all links with <a> tag and print links with Href
```

```python
AllLinks= soup.find_all('a')
print(AllLinks)
print()
for i in AllLinks:
    print(i.get("href"))
```

```python
#Extracting only thirty table rows data
```

```python
TableRows=soup.find_all('tr')
AllRows=TableRows[:30]
print(AllRows)
```

```python
#parsing string and printing as Dataframe
```

```python
import re

list_rows = []
for row in AllRows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
DF=pd.DataFrame(list_rows)
print(DF.head(20))
```

```python
#Find header and clean them
```

```python
headers=soup.find_all("th")
print(headers)
```

```python
import re
regex = re.compile(r'<[^>]+>')
Header=[]
for i in headers:
    strr=str(i)
    result=regex.sub('',strr)
    Header.append(result)
print(Header)
```
